```python
from astropy.io import fits
from astropy.visualization import PercentileInterval, AsinhStretch, MinMaxInterval, SqrtStretch, ImageNormalize, SinhStretch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# OPEN FITS IMAGES AND INSPECT META DATA

file_path = file_paths[2]
quad = 2

hdu_1 = fits.open(file_path)
info = hdu_1.info()
image_data = hdu_1[quad].data
header = hdu_1[quad].header
prime_header = hdu_1[0].header
hdu_1.close()

# VISUALIZE THE DATA

# Scale the image based on its percentiles
interval = PercentileInterval(99.5)
vmin, vmax = interval.get_limits(image_data)
norm = plt.Normalize(vmin, vmax)

# Apply an asinh stretch to enhance faint features
stretch = AsinhStretch()
scaled_data = stretch(norm(image_data))

# Plot the image
plt.figure(figsize = (15,15))
plt.imshow(scaled_data, cmap='gray')
plt.show()

# primary header card
for card in prime_header.cards:
    print(card)

print('\n\n')

# image header
for card in header.cards:
    print(card)
```
